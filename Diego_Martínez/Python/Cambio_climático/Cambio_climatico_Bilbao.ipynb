#Creo un bucle para acceder a la API de aemet Open Data y descargo la info de los datos y estaciones meteorol√≥gicas deseadas

```python
import requests
import pandas as pd

listaano= []

#ano = "1973"

# Guardo el periodo de a√±os en una lista como string
for ano in range(2014,2024,1):
  ano = str(ano)
  listaano.append(ano)

# Bucle que guarda la info de cada a√±o
for ano in listaano:

  url = "https://opendata.aemet.es/opendata/api/valores/climatologicos/mensualesanuales/datos/anioini/"+ano+"/aniofin/"+ano+"/estacion/1082"

  querystring = {"api_key":"eyJhbGciOiJIUzI1NiJ9.eyJzdWIiOiJtYXJ0aW5lenBhcmRvZGllZ29AZ21haWwuY29tIiwianRpIjoiYTU1YjczNDktYjIxOS00MWRiLTg5YzktYzRiNWU1NWI3OTdmIiwiaXNzIjoiQUVNRVQiLCJpYXQiOjE3MTQ5OTc0NzAsInVzZXJJZCI6ImE1NWI3MzQ5LWIyMTktNDFkYi04OWM5LWM0YjVlNTViNzk3ZiIsInJvbGUiOiIifQ.PaduQ1vHGQCTaWPEhsYpmFKSC1jrGoa6H8dWDzKqNYs"}

  headers = {
      'cache-control': "no-cache"
      }

  response = requests.request("GET", url, headers=headers, params=querystring)
  valor = response.json()
  endpoint = valor['datos']


  df = pd.read_json(endpoint)


  clima = pd.concat([clima, df], axis=0)      #Concatenamos los datos

clima
```
