# **Búsqueda sin información**
## El paisano, el lobo, el cordero y el heno quieren pasar el río en un bote donde solo caben dos elementos

## En ausencia del paisano, el lobo se come al cordero y el cordero se como el heno,

### **El paisano quiere pasarlos al otro lado del río sin perder nada**

Necesitamos varias cosas:
 - Definir lo que es un estado, además de definir los estados iniciales y finales.
 - Una función que a partir de un estado cree todos los estados posibles a partir de él.
 - Una función que dado un estado nos diga si es factible.
 - Una función que dado un estado nos diga si es la solución.

```python
import copy;                                       #la usamos para hacer copias de los estados


estadoInicial = [{"P","L","C","H"},set()];          #P paisano, L lobo, C cordero, H heno son dos conjuntos uno los animales, dos un conjunto vacio set, la otra orilla

estadoFinal = [set(), {"P","L","C","H"}];

def esFactible(estado, lado=None):
    if lado == None:
       return esFactible(estado, 0) and esFactible(estado,1); #comprueba si el estado de la izq. es factible y el de la derecha
    else:
       if "P" in estado[lado]: return True; #Si el paisano está en ese lado true
       if "L" in estado[lado] and "C" in estado[lado]: return False;
       if "H" in estado[lado] and "C" in estado[lado]: return False;
       return True; #El resto de convinaciones son correctas

#función que devuelve los estados posibles a partir de un estado dado

def movimientos(estado):
    lp = (0 if "P" in estado[0] else 1);
    np = (1 if "P" in estado[0] else 0);

    rt = [];

    #primer movimiento, paso que solo movemos al paisano
    ns = copy.deepcopy(estado);     #ns; nuevo estado ; np, no está el paisano; lp, lado del paisano
    ns[np] = ns[np].union({"P"});   # añadp P al nuevo estado
    ns[lp] = ns[lp].difference({"P"});  # elimino P del otro estado

    rt.append(ns);

    #convinaciones que pueden ir con el paisano
    for c in estado[lp].difference({"P"}):      #en el lado que está el paisano, quita al paisano
        ns=copy.deepcopy(estado);
        ns[np] = ns[np].union({"P",c});         #añado el nuevo estado cambiando al paisano de lado con un compañero c
        ns[lp] = ns[lp].difference({"P",c});    #elimino al paisano y al compañero c
        rt.append(ns)

    return rt;

def esSolución(estado):
    return (estado == estadoFinal);

#movimientos(estadoInicial)


```

```python
# algoritmo general de búsqueda en un espacio de estados

abiertos = [estadoInicial];
cerrados = [];

actual = abiertos[0];
while not esSolución(actual) and not len(abiertos)==0:           #desde el estado inicial va añadiendo estados en cerrados
      abiertos.pop(0);
      cerrados.append(actual);
      hijos=movimientos(actual);
      for hijo in hijos:                                          #es factible y no está en actual la añade
          if esFactible(hijo) and not hijo in cerrados:
             abiertos.append(hijo);
      actual=abiertos[0];

print(cerrados)
```

```python
# algoritmo general de búsqueda en profundidad en un espacio de estados usando recursividad

def búsquedaEnProfundidad(actual,cerrados,solución):          #función que voy a llamar desde el estado inicial y va ir generando todo el árbol y busca en profundidad
    if esSolución(actual):
       solución.append(actual)               #compruebo si el estado que me pasan es la solución, si es así la añadimos a solución
       return True;                          #si es la solución devuelvo un true
    else:                                    #si no, genero los movimiento hijos y para cada hijo comprueba si es factible usando hijo como el actual
       hijos=movimientos(actual);
       for hijo in hijos:
           if esFactible(hijo) and not hijo in cerrados:
              cerrados.append(hijo);
              r=búsquedaEnProfundidad(hijo,cerrados,solución);
              if r==True:
                 solución.append(actual);
                 return True;
       return False;

solución=[];
búsquedaEnProfundidad(estadoInicial,[],solución);
print(solución); #de la solución final muestra cada uno de los pasos correctos hasta llegar a la solución inicial, si lo quiero ver al revés, usamos solución.reverse
```
