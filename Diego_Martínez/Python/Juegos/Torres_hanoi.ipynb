```python
import copy;                             #la usamos para hacer copias de los estados
import random;


estadoInicial = [[3,2,1],[],[]];          #Un poste con los 3 disco oredenados

estadoFinal = [[],[],[3,2,1]];            #Los tres discos ordenados en el último poste


def selectTorre():           #Seleccionamos dos torres de las tres al azar, la torre origen donde cogemos el disco y la torre destino donde lo vamos a colocar
    torreOrigen = random.randrange(3);
    torreDestino = random.randrange(3);
    return torreOrigen, torreDestino;           #Devolvemos las dos torres seleccionadas


def esFactible(torreDestino,torreOrigen): #Comprobamos que el movimiento es factible

    if len(torreOrigen) == 0:             #Si la torre origen está vacia no puede seleccionar un disco
        return False;

    if len(torreDestino) == 0:            #Si en la torre destino no hay discos se puede mover
        return True;



    if torreOrigen[-1] >= torreDestino[-1]: #Si el último disco de la torre origen es mayor o igual que el último disco de la torre destino no se puede mover
        return False;

    return True;                          #el resto de movimientos son correctos


def moverDisco(torreOrigen,torreDestino):

    torreDestino.append(torreOrigen[-1]);
    torreOrigen.remove(torreOrigen[-1]);


def esSolucion(estado):

    return (estado == estadoFinal);



```

```python
#Busca la solución de forma aleatoria, le pongo un tope para que finalice la ejecución si no encuentra la solución antes

jugadas = 0;
topeJugadas = 150;

hanoi = copy.deepcopy(estadoInicial);
finJuego = False;

print("Empezamos el juego: {}".format(hanoi));

while not esSolucion(hanoi) and jugadas < topeJugadas:
      torreOrigen, torreDestino = selectTorre();

      if esFactible(hanoi[torreDestino],hanoi[torreOrigen]) == True:
          moverDisco(hanoi[torreOrigen], hanoi[torreDestino]);
          print("jugada {}: {}".format(jugadas+1, hanoi));
          jugadas += 1;

      if esSolucion(hanoi) == True:
          finJuego = True;
          print("Juego finalizado con solucción optima");

if finJuego == False:
    print("Fin de movimientos sin solución optima")
```
